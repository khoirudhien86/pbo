```python
#Jawab:
#saya mendapat tugas untuk membuat program tentang prisma segitiga pada uts kemarin.
#lalu di uas ini saya membuat program tentang prisma segitiga.

class VolumePrismaSegitiga:
    def __init__(self):
        print("MENGHITUNG VOLUME PRISMA SEGITIGA")
        print("=================================")
        print()
        self.__tinggiPrisma = int(input("masukkan tinggi prisma = "))
        self.__alasSegitiga = int(input("masukkan alas segitiga = "))
        self.__tinggiSegitiga = int(input("masukkan tinggi segitiga = "))
        
    def menghitungVolume(self):
        volume = (0.5*self.__alasSegitiga*self.__tinggiSegitiga)*self.__tinggiPrisma
        print()
        print("--------------------------------------------")
        print("Volume Prisma Segitiga nya adalah : ",volume)
        
class MenghitungBangunDatar:
    def __init__(self):
        pass
        
    def menghitungAlasPrisma(self):
        self.__tinggiSegitiga = int(input("masukkan tinggi segitiga = "))
        self.__alasSegitiga = int(input("masukkan alas segitiga = "))
        luasAlas = (0.5*self.__alasSegitiga*self.__tinggiSegitiga)
        print("----------------------------------------------")
        print("Luas alas prisma Segitiga nya adalah : ",luasAlas)
        print("----------------------------------------------")
        return luasAlas
        
    def menghitungKelilingAlas(self):
        print()
        self.__sisi1 = int(input("masukkan sisi 1 = "))
        self.__sisi2 = int(input("masukkan sisi 2 = "))
        self.__sisi3 = int(input("masukkan sisi 3 = "))
        kelilingAlas = self.__sisi1+self.__sisi2+self.__sisi3
        print("-----------------------------------------------")
        print("Keliling alas prisma Segitiga nya adalah : ",kelilingAlas)
        print("-----------------------------------------------")
        return kelilingAlas


class LuasPermukaanPrisma(MenghitungBangunDatar):
    def __init__(self):
        print("MENGHITUNG LUAS PERMUKAAN PRISMA SEGITIGA")
        print("=========================================")
        print()
        super().__init__()
        self.__tinggiPrisma = int(input("masukkan tinggi prisma = "))
        self.luasAlas = self.menghitungAlasPrisma()
        self.kelilingAlas = self.menghitungKelilingAlas()
        
    def menghitungLuasPrisma(self):
        Luas = (2*self.luasAlas)+(self.kelilingAlas*self.__tinggiPrisma)
        print()
        print("===================================================")
        print("Luas Permukaan Prisma Segitiga nya adalah : ",Luas)
        print("===================================================")
```
